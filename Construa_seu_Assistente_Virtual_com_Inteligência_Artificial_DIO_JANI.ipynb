```python
# ============================================
# ASSISTENTE VIRTUAL COM IA - DESAFIO DIO
# 100% ONLINE - Google Colab
# ============================================

# 1. INSTALAR BIBLIOTECAS
!pip install -q gTTS SpeechRecognition pydub wikipedia

# 2. IMPORTAÇÕES
import speech_recognition as sr
from gtts import gTTS
import playsound
import os
import webbrowser
import wikipedia
from pydub import AudioSegment
import IPython.display as ipd
import base64
from io import BytesIO

print("="*60)
print("🤖 ASSISTENTE VIRTUAL COM IA - DIO")
print("="*60)

# 3. FUNÇÃO PARA FALAR (TTS)
def falar(texto):
    """Converte texto em áudio e reproduz no Colab"""
    print(f"🗣️ Assistente: {texto}")
    tts = gTTS(text=texto, lang='pt-br')
    # Salva em memória para reproduzir no Colab
    fp = BytesIO()
    tts.write_to_fp(fp)
    fp.seek(0)
    audio_base64 = base64.b64encode(fp.read()).decode()
    display(ipd.Audio(data=base64.b64decode(audio_base64), autoplay=True))

# 4. FUNÇÃO PARA OUVIR (STT)
def ouvir():
    """Escuta o microfone e transcreve a fala"""
    recognizer = sr.Recognizer()

    # No Colab, usamos o microfone do navegador
    # Para este exemplo, vamos simular com entrada de texto
    # (explicarei como usar voz real depois)

    print("\n🎤 Digite seu comando (ou 'sair' para encerrar):")
    comando = input(">>> ")
    return comando.lower()

# 5. PROCESSADOR DE COMANDOS
def processar_comando(comando):
    """Processa o comando do usuário e executa ações"""

    if "sair" in comando or "tchau" in comando:
        falar("Até logo! Foi um prazer ajudar você.")
        return False

    elif "olá" in comando or "oi" in comando:
        falar("Olá! Como posso ajudar você hoje?")

    elif "pesquisar" in comando or "wikipedia" in comando:
        try:
            # Extrai o termo de pesquisa
            termo = comando.replace("pesquisar", "").replace("wikipedia", "").strip()
            if not termo:
                falar("O que você gostaria de pesquisar?")
                return True

            falar(f"Pesquisando sobre {termo} na Wikipedia...")
            resultado = wikipedia.summary(termo, sentences=3, auto_suggest=False)
            falar(resultado[:500])  # Limita para não falar demais

            # Mostra também o link
            print(f"🔗 Link: https://pt.wikipedia.org/wiki/{termo.replace(' ', '_')}")

        except wikipedia.exceptions.PageError:
            falar(f"Não encontrei nada sobre {termo} na Wikipedia.")
        except wikipedia.exceptions.DisambiguationError:
            falar(f"Encontrei várias opções para {termo}. Seja mais específico.")
        except Exception as e:
            falar("Desculpe, tive um problema ao pesquisar.")
            print(f"Erro: {e}")

    elif "youtube" in comando or "vídeo" in comando:
        falar("Abrindo o YouTube...")
        webbrowser.open("https://www.youtube.com")
        print("🔗 YouTube aberto no navegador!")

    elif "farmácia" in comando or "mapa" in comando:
        falar("Abrindo Google Maps para encontrar farmácias próximas...")
        webbrowser.open("https://www.google.com/maps/search/farmacia+proxima")
        print("🔗 Google Maps aberto!")

    elif "ajuda" in comando:
        falar("Posso ajudar você com os seguintes comandos:")
        print("""
        📋 COMANDOS DISPONÍVEIS:
        - "Olá" ou "Oi" → Saudação
        - "Pesquisar [assunto]" → Busca na Wikipedia
        - "YouTube" → Abre o YouTube
        - "Farmácia" → Busca farmácias no mapa
        - "Ajuda" → Mostra esta lista
        - "Sair" ou "Tchau" → Encerra
        """)

    else:
        falar("Não entendi seu comando. Diga 'ajuda' para ver o que posso fazer.")

    return True

# 6. LOOP PRINCIPAL
def main():
    print("\n🎯 Assistente Virtual iniciado!")
    falar("Olá! Sou seu assistente virtual. Como posso ajudar?")
    print("\n💡 Digite 'ajuda' para ver os comandos disponíveis")
    print("💡 Digite 'sair' para encerrar")

    continuar = True
    while continuar:
        comando = ouvir()
        if comando:
            continuar = processar_comando(comando)
            print("-"*50)

# 7. EXECUTAR
if __name__ == "__main__":
    main()
```
